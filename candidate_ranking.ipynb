```python
honeypot_scores = []

for r in records:

    penalty = 0

    years = r["profile"]["years_of_experience"]

    total_months = sum(
        job["duration_months"]
        for job in r["career_history"]
    )

    # Career duration impossible
    if total_months / 12 > years + 2:
        penalty += 2

    # Too many advanced skills
    advanced = sum(
        s["proficiency"] == "advanced"
        for s in r["skills"]
    )

    if advanced >= 10:
        penalty += 2

    # Too many skills
    if len(r["skills"]) > 25:
        penalty += 1

    honeypot_scores.append(
        (
            r["candidate_id"],
            penalty
        )
    )
```

```python
honeypot_df = pd.DataFrame(
    honeypot_scores,
    columns=[
        "candidate_id",
        "honeypot_penalty"
    ]
)
```

```python
reasonings = []

for cid in top100["candidate_id"]:

    facts = candidate_facts[cid]
    techs = matched_tech[cid]

    # ---------- Strength ----------
    strength = []

    strength.append(
        f"{facts['experience']:.1f} years of experience as a {facts['title']}."
    )

    if len(techs) > 0:
        strength.append(
            "Hands-on work in " + ", ".join(techs[:3]) + "."
        )

    # ---------- Positive Signals ----------
    positives = []

    if facts["response"] >= 0.80:
        positives.append("Strong recruiter engagement.")

    if facts["github"] >= 60:
        positives.append("High GitHub activity.")

    if facts["open"]:
        positives.append("Open to work.")

    # ---------- Concern ----------
    concern = ""

    if facts["notice"] >= 90:
        concern = "Long notice period may delay onboarding."

    elif facts["notice"] >= 60:
        concern = "Moderate notice period."

    elif not facts["open"]:
        concern = "Currently not marked open to work."

    # ---------- Final Reason ----------
    reason = " ".join(strength)

    if positives:
        reason += " " + " ".join(positives)

    if concern:
        reason += " " + concern

    reasonings.append(
        (
            cid,
            reason
        )
    )
```

```python
import re

def get_best_sentence(record):

    priority_words = {
        "improved":5,
        "built":4,
        "developed":4,
        "implemented":4,
        "designed":3,
        "trained":3,
        "deployed":3,
        "created":3,
        "owned":2,
        "recommendation":3,
        "ranking":3,
        "semantic search":3,
        "retrieval":3,
        "rag":3,
        "faiss":3,
        "pinecone":3,
        "bm25":3,
        "bge":3
    }

    best_sentence = ""
    best_score = -1

    for job in record["career_history"]:

        sentences = re.split(r'(?<=[.!?])\s+', job["description"])

        for sentence in sentences:

            s = sentence.lower()

            score = 0

            # important keywords
            for word, wt in priority_words.items():
                if word in s:
                    score += wt

            # numbers = impact
            if "%" in sentence:
                score += 5

            if re.search(r"\d+\s*M", sentence):
                score += 5

            if re.search(r"\d+\s*K", sentence):
                score += 4

            if len(sentence) > 250:
                score -= 2

            if score > best_score:
                best_score = score
                best_sentence = sentence.strip()

    return best_sentence
```

```python
def detect_project(text):

    text = text.lower()

    if "recommendation" in text:
        return "recommendation"

    elif "semantic search" in text:
        return "semantic_search"

    elif "rag" in text:
        return "rag"

    elif "ranking" in text:
        return "ranking"

    elif "retrieval" in text:
        return "retrieval"

    return "general"
```

```python
def extract_tech(text):

    techs = []

    keywords = [
        "FAISS",
        "Pinecone",
        "Milvus",
        "BM25",
        "BGE",
        "Sentence Transformer",
        "OpenAI",
        "GPT-4",
        "LoRA",
        "QLoRA",
        "LightGBM",
        "XGBoost"
    ]

    lower = text.lower()

    for tech in keywords:

        if tech.lower() in lower:
            techs.append(tech)

    return techs
```

```python
def summarize_project(sentence):

    s = sentence.lower()

    if "pinecone" in s and "openai" in s:
        return "Built a production RAG pipeline using OpenAI embeddings and Pinecone for enterprise knowledge retrieval."

    elif "recommendation" in s:
        return "Built large-scale recommendation systems with production ranking pipelines."

    elif "semantic search" in s:
        return "Developed semantic search solutions using dense embeddings and vector search."

    elif "ranking" in s:
        return "Designed production ranking pipelines for search and recommendation systems."

    elif "retrieval" in s:
        return "Implemented dense retrieval pipelines for large-scale search systems."

    return sentence
```

```python
def extract_technologies(text):

    techs = [
        "FAISS",
        "Pinecone",
        "Milvus",
        "BM25",
        "Semantic Search",
        "Recommendation",
        "Ranking",
        "RAG",
        "Embeddings",
        "BGE",
        "Sentence Transformers",
        "XGBoost",
        "LightGBM",
        "LoRA",
        "QLoRA"
    ]

    found = []

    lower = text.lower()

    for tech in techs:

        if tech.lower() in lower:
            found.append(tech)

    return found
```

```python
fit_templates = [
    "Closely matches the technical requirements of the target role.",
    "Demonstrates strong production experience relevant to the job description.",
    "Career history aligns well with the search, retrieval and recommendation requirements.",
    "Previous projects closely match the responsibilities of this position.",
    "Technical background is highly relevant for production AI systems.",
    "Shows practical experience building large-scale ML and retrieval solutions."
]

reasonings = []

for cid in top100["candidate_id"]:

    r = record_map[cid]
    signal = r["redrob_signals"]

    intro = (
        f"{r['profile']['current_title']} with "
        f"{r['profile']['years_of_experience']:.1f} years of experience."
    )

    achievement = get_best_sentence(r)

    candidate_number = int(cid.split("_")[1])

    fit = fit_templates[
        candidate_number % len(fit_templates)
    ]




    positives = []

    if signal["github_activity_score"] >= 60:
      positives.append("High GitHub activity indicates continued technical engagement.")

    if signal["recruiter_response_rate"] >= 0.80:
      positives.append("High recruiter response suggests strong market interest.")
    if signal["open_to_work_flag"]:
      positives.append("Currently open to new opportunities.")

    concern = ""

    if signal["notice_period_days"] >= 90:
        concern = "Long notice period may delay onboarding."

    elif signal["notice_period_days"] >= 60:
        concern = "60-day notice period."

    elif not signal["open_to_work_flag"]:
        concern = "Currently not marked open to work."

    reasoning = (
        intro
        + " "
        + achievement
        + " "
        + fit
    )

    if positives:
        reasoning += " " + " ".join(positives)

    if concern:
        reasoning += " " + concern

    reasonings.append((cid, reasoning))
```
